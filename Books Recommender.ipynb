```python
def recommend_book(book_name):
    # Check if the book_name exists in book_pivot.index
    if book_name not in book_pivot.index:
        print(f"Book '{book_name}' not found in the dataset.")
        return

    # Proceed with the recommendation if the book is found
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    # Print the suggestions
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched for '{book_name}'\n")
                print("The suggested books are: \n")
            else:
                print(j)

```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
